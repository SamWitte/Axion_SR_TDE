Comparison between n=3 and n=4, Non-relativistic!

```python
# fa12 = np.loadtxt("../test_store/OutRun_Lvln4_NonRel_fa_1.0e12_ma_2.0e-12_MBH_22.2_spin_0.99_.dat")
# fa12Lvl4 = np.loadtxt("../test_store/OutRun_Lvln4_FullRel_fa_1.0e12_ma_2.0e-12_MBH_22.2_spin_0.99_.dat")
fa12Lvl4 = np.loadtxt("../test_store/OutRun_Lvln4_NonRel_fa_1.0e12_ma_2.0e-12_MBH_22.2_spin_0.99_v2_.dat")
fa12 = np.loadtxt("../test_store/OutRun_NonRel_fa_1.0e12_ma_2.0e-12_MBH_22.2_spin_0.99_v2_.dat")
```

```python
fig, axs = plt.subplots(2, sharex=True,  figsize=(10,6), gridspec_kw={'hspace': 0.07, 'height_ratios': [1, 3]})

lww=1.1

axs[1].plot(fa12[:,0], fa12[:,1], "#6F1D1B", lw=lww,  dashes=[4, 4])
axs[1].plot(fa12[:,0], fa12[:,2], "#031D44", lw=lww, dashes=[4, 4])
#axs[1].plot(fa12[:,0], fa12[:,3], "#00A676", lw=lww,  dashes=[4, 4])
#axs[1].plot(fa12[:,0], fa12[:,4], "#B68F40", lw=lww, dashes=[4, 4])
#axs[1].plot(fa12[:,0], fa12[:,5], "#6DAEDB", lw=lww,  dashes=[4, 4])


axs[1].plot(fa12Lvl4[:,0], fa12Lvl4[:,1], "#6F1D1B", lw=lww)
axs[1].plot(fa12Lvl4[:,0], fa12Lvl4[:,2], "#031D44", lw=lww)
axs[1].plot(fa12Lvl4[:,0], fa12Lvl4[:,3], "#00A676", lw=lww)
axs[1].plot(fa12Lvl4[:,0], fa12Lvl4[:,4], "#B68F40", lw=lww)
axs[1].plot(fa12Lvl4[:,0], fa12Lvl4[:,5], "#6DAEDB", lw=lww)

axs[0].plot(fa12[:,0], fa12[:,-2], "k", dashes=[4, 4], lw=lww)
axs[0].plot(fa12Lvl4[:,0], fa12Lvl4[:,-2], "k", lw=lww)


axs[1].axhline(7.046e-11, alpha=0.5, lw=1, ls ="dotted", c="k")
axs[1].axhline(7.046e-11 * (3/2)**4, alpha=0.5, lw=lww, ls ="dotted", c="k")
axs[1].axhline(7.046e-11 * (4/2)**4, alpha=0.5, lw=lww, ls ="dotted", c="k")

axs[1].fill_between(fa12Lvl4[:,0], np.ones(len(fa12Lvl4[:,0])) * 7.046e-11 * (4/2)**4, 1e100, color="k",ec="k", alpha=0.1)


axs[1].set_yscale("log")
axs[0].set_yscale("linear")
axs[1].set_xscale("log")
axs[1].set_ylim([1e-18,1e-7])
# axs[1].set_xlim([10,5e6])
axs[1].set_xlim([5e-3,4e6])
axs[0].set_ylim([-0.05,1.2])
axs[0].set_ylabel(r"$\tilde{a}$", fontsize=23)
axs[1].set_ylabel(r"$\epsilon_x$", fontsize=23)
axs[1].set_xlabel(r"t   [yrs]", fontsize=23);
axs[1].text(1e4, 1e-15, r"$f_a = 10^{12} \, {\rm GeV}$", fontsize=18);

axs[1].text(1e-2, 7e-9, r"Bosenova Threhsold", fontsize=15);

axs[1].xaxis.set_minor_locator(mpl.ticker.LogLocator(subs="all", numticks=90))
axs[1].xaxis.set_major_locator(mpl.ticker.LogLocator(numticks=90))

plt.savefig("Solve_f12.jpeg", dpi=200)
```
